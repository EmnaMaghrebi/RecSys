```python
ratings_df = pd.read_csv('ratings.csv')
ratings_df.head()
```

```python
ratings_df = ratings_df.groupby('user_hash').max()
ratings_df.head()
```

```python
ratings_df.dropna(inplace=True, how='all')
ratings_df.replace(np.nan, 0, inplace=True)
ratings_df.head()
```

```python
ratings_df.to_csv('ratings_per_user.csv')
```

```python
ratings_df['Group'] = ratings_df.index.map(ratings)
ratings_df.dropna(inplace=True)
len(ratings_df)
```

```python
splitted_dataframes = {}
# split dataframes by group
for group in ratings_df['Group'].unique():
    group_df = ratings_df.loc[ratings_df['Group'] == group].copy()
    group_df.drop(columns=['Group'], inplace=True)
    group_df = group_df.loc[:, (group_df != 0).any(axis=0)]
    splitted_dataframes[group] = group_df
```

```python
group_0_df = splitted_dataframes[0]
group_1_df = splitted_dataframes[1]
group_2_df = splitted_dataframes[2]
group_3_df = splitted_dataframes[3]
group_4_df = splitted_dataframes[4]
```
