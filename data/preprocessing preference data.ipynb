```python
preference_df = pd.read_csv('preferences.csv')
preference_df.head()
```

Extract items from the column names

```python
items = []

for col_name in preference_df.columns:
    tuple_items = ast.literal_eval(col_name)
    items.append(tuple_items)
print(items[0])
# Remove duplicates by converting the list to a set
n_items = len(set(items))
n_items
```

```python
items_tuples = []

for col_name in preference_df.columns:
    tuple_items = ast.literal_eval(col_name)
    items_tuples.append(tuple_items)

# Remove duplicates by converting the list to a set
n_items = len(set(items_tuples))

# Get the list of all unique items
items = set(item for pair in items_tuples for item in pair)
# Map items to integer indices
item_to_idx = {item: idx for idx, item in enumerate(items)}

# Initialize 3D tensor
n_users = len(preference_df)
n_items = len(items)
Y = np.zeros((n_users, n_items, n_items))
# Fill in tensor
for i, user in enumerate(preference_df.index):
    for col_name, score in preference_df.loc[user].items():
        item1, item2 = ast.literal_eval(col_name)
        j, k = item_to_idx[item1], item_to_idx[item2]
        if score < 0:
            Y[i, j, k] = -1
        elif score > 0:
            Y[i, j, k] = 1
        # If score == 0 or comparison not present, Y[i, j, k] remains 0
```
