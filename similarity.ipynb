```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import scipy.stats as stats
```

```python

# standard deviation of spherical Gaussian distribution
mu = 0.0        #zero-mean
sigma_u = 1.0
sigma_v = 1.0
sigma = 1.0 #(sigma^2 = 0.1, 0.3, 1.0, 3, 10)

# Generate U, V from the Gaussian distribution
# U : D*N
# V : D*M

D = 2 #dimension
N = 1000 #nbr of users = 10
M = 500 #nbr of movies : M=5*alpha
alpha=[1,5,10,50,100]

U = np.random.normal(mu, sigma_u, size=(D,N))
V = np.random.normal(mu, sigma_v, size=(D,M))
noise = np.random.normal(mu, sigma, size=(N,M))
R = np.matmul(U.T,V) + noise
# Print ratings matrix R
print(U.shape)
```

```python
df = pd.DataFrame(R)
df.head()
```

```python
def condition(x):
    if x<-2:
        return 1
    elif x>=-2 and x<=-0.5:
        return 2
    elif x>-0.5 and x<=0.5:
        return 3
    elif x>0.5 and x<=2:
        return 4
    else:
        return 5
```

```python
df2 = df.applymap(condition)
df2.head()
```

```python
#similarity over matrix R
#similarity over matrix U
```

```python
#add a column containing first row of sim.matrix , sort using this column to obtain users in order of similarity, 
#then drop col. to compare later with the other sim. matrix(U)
sim_array= cosine_similarity(df2)[0]

df2['similarity_u1']= sim_array.tolist()
df2.head()
```

```python
df2= df2.sort_values(by='similarity_u1', ascending= False)
df2.head(10)
#sort_df= df2.drop('similarity_u1', axis=columns)
```
